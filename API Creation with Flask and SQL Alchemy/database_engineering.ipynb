```python
import sqlalchemy
import pymysql
from sqlalchemy import create_engine
from sqlalchemy.ext.declarative import declarative_base
from sqlalchemy import Column, Integer, String, Float
from sqlalchemy.orm import Session
import pandas as pd
import sqlite3
import csv
```

```python
Base = declarative_base()
```

```python
class Measurement(Base):
    __tablename__ = 'measurement'
    index = Column(Integer, primary_key = True, default = 0)
    station = Column(String(255))
    date = Column(String(255))
    prcp = Column(Float)
    tobs = Column(Float)
```

```python
class Station(Base):
    __tablename__ = 'station'
    index = Column(Integer, primary_key = True, default = 0)
    station = Column(String(255))
    name = Column(String(255))
    latitude = Column(Float)
    longitude = Column(Float)
    elevation = Column(Float)
```

```python
measurement_df = pd.read_csv('clean_hawaii_measurements.csv')
measurement_df.head()
```

```python
station_df = pd.read_csv('hawaii_stations.csv')
station_df.head()
```

```python
session = Session(bind=engine)
conn = sqlite3.connect('hawaii.sqlite')
cur = conn.cursor()
```

```python
with open('clean_hawaii_measurements.csv') as csvfile:
    reader = csv.reader(csvfile)
    for row in reader:
        cur.execute('INSERT INTO Measurement VALUES (NULL,?,?,?,?);', row)
```

```python
#iterating through length of station dataframe to add data into sqlite table
for i in range(len(station_df)):
   session.add(Station(index = station_df.index[i], station = station_df.iloc[i][0], name = station_df.iloc[i][1], latitude = station_df.iloc[i][2], longitude = station_df.iloc[i][3], elevation = station_df.iloc[i][4]))
```

```python
#same for measurement data
for i in range(len(measurement_df)):
   session.add(Measurement(index = measurement_df.index[i], station = measurement_df.iloc[i][0], date = measurement_df.iloc[i][1],
                           prcp = measurement_df.iloc[i][2], tobs = int(measurement_df.iloc[i][3])))
```

```python
query1 = session.query(Station)
for row in query1:
    print(row.station)
```
